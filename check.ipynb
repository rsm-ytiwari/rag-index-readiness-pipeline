```python
import json
import pandas as pd

print("Loading first 100,000 reviews...")

reviews = []
with open('./Yelp JSON/yelp_dataset/yelp_academic_dataset_review.json', 'r', encoding='utf-8') as f:
    for i, line in enumerate(f):
        if i >= 7000000:
            break
        reviews.append(json.loads(line))

        # Progress indicator
        if (i + 1) % 10000 == 0:
            print(f"  Loaded {i+1:,} reviews...")

df = pd.DataFrame(reviews)

print(f"\n{'='*50}")
print(f"Successfully loaded {len(df):,} reviews!")
print(f"{'='*50}")

print("\nDataFrame info:")
print(df.info())

print("\nFirst review:")
print(df.iloc[0])

print("\nColumns available:")
print(df.columns.tolist())
```
